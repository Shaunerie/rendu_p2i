## I. Présentation générale
---
<a id="Ia"></a>
### I.a) Résumé du projet :

<p style="text-align: justify;"> <b>Les matériaux granulaires représentent un enjeu majeur de notre société</b>; ils sont présents dans tous les domaines sous différentes formes, du sable à la roche en passant par le blé. Comprendre leur comportement dans les milieux qui les entourent est donc essentiel pour faire des avancées dans de nombreux secteurs (géologie, industries minières/pharmaceutique/chimique...) mais également pour optimiser leur transport et leur stockage. </p>

<p style="text-align: justify;"> Dans notre projet, nous nous sommes concentrés sur <b>le secteur de l'agroalimentaire avec les silos à grains.</b> L'objectif principal du projet est donc de créer une simulation, en python, permettant de <b>vérifier numériquement une formule analytique du débit granulaire</b> à la sortie d'un silo à grain. </p>

<a id="Ib"></a>
### I.b) Brève introduction à la modélisation numérique :
<p style="text-align: justify;"> Afin de réaliser cette modélisation, nous utilisons une méthode de discrétisation appelée <b>Discrete Element Method (DEM)</b>. Cette méthode permet de simuler les interactions entre les grains et les parois du silo en considérant chaque grain comme une entité discrète. </p>

<p style="text-align: justify;"> Chaque grain, possède sa propre taille et masse. A chaque instant on associe une position, une vitesse et une accélération à tous les grains de notre simulation. </p>


<p style="text-align: justify;"> Pour réaliser la simulation il nous faut aussi <b>discrétiser le temps.</b> On définit une durée pour la simulation ainsi qu'un pas de temps déterminant l'intervalle d'actualisation des caractéristiques de nos grains. Autrement dit, <b>à tous les pas de temps, nous mettons à jour les intéractions sur les grains</b> afin de calculer leur nouvelle position, vitesse et accélération. </p>

## II. Démarche scientifique 
---

<p style="text-align: justify;"> Pour élaborer notre <b>approche scientifique</b> et les <b>méthodes numériques</b> à utiliser, nous avons principalement utilisé le cours et les ressources fournies par notre tuteur. </p>

<a id="IIa"></a>
### II.a) Contacts et paramètres physiques :

<p style="text-align: justify;"> Parmi ces méthodes, on retrouve <b>la modélisation des contacts</b> entre les différents éléments de notre simulation. </p>

<p style="text-align: justify;"> En vue de modéliser un contact entre deux entités, on introduit des <b>ressorts</b> ainsi qu'un <b>amortisseur</b> entre les deux grains. Cette approche permet de mettre en oeuvre <b>les efforts répulsifs et attractifs</b> qui découlent d'un contact entre deux objets. Ci-dessous un schéma descriptif: </p>
<a id="figure1"></a>
<figure style="text-align: center; padding: 20px">
    <img src=ressort.png align="center" width=600>
    <figcaption style="border: 1px solid black; padding: 10px; display: inline-block;"><b>Fig.1 - Représentation schématique du contact entre deux grains</b></figcaption>
</figure>

<p style="text-align: justify;"> Les valeurs $ \delta_n $ et $\delta_t $ représentent respectivement <b>l'allongement normal et tangentiel</b>. Ils représentent <b>l'interpénétration</b> des deux grains. Plus les deux objets s'interpénetrent plus les efforts repulsifs sont importants. Nous pouvons faire l'analogie avec un ressort: plus on compresse un ressort plus la réaction du ressort est importante </p>

<p style="text-align: justify;"> Les paramètres $k_n, \gamma_n, \mu $ sont des constantes définissant respectivement <b>la raideur, le coefficient d'amortissement et le coefficient de frottement.</b></p>

<p style="text-align: justify;"> Ces paramètres permettent de caractériser et de modifier manuellement les intéractions qui ont lieu. En augmentant la raideur, par exemple, nous amplifions la réaction du ressort proportionnellement à son allongement. </p>
    
<a id="ref_relation"></a>
<p style="text-align: justify;"> <b>Le choix de toutes ces valeurs est crucial pour le bon fonctionnement de la simulation</b>. En effet, si les paramètres sont incohérents et/ou incohérents entre eux, bien que le code soit juste, la simulation sera fausse. Le choix de ces valeurs est <b>arbitraire</b> et doit être changé <b>manuellement</b> pour s'adapter à chaque modélisation. Cependant, il existe des <b>formules usuelles</b> reliant ces grandeurs entre-elles fournissant des résultats satisfaisant dans la plupart des temps. </p>

<p style="text-align: justify;"> Prenons l'exemple de l'effort normal au contact entre deux grains : </p>
    
<a id="figure2"></a>
<figure style="text-align: center; padding: 20px">
    <img src="equation_ressort.png" alt="Résultante de l'effort normal" width="600">
    <figcaption style="border: 1px solid black; padding: 10px; display: inline-block;"><b>Fig.2 - Résultante de l'effort normal</b></figcaption>
</figure>



<p style="text-align: justify;"> Le terme $k_n \cdot \delta_n$ représente <b>l'effort répulsif</b> lorsque deux grains s'interpénètrent, c'est le terme analogique à la résultante d'un ressort. </p>

<p style="text-align: justify;"> Le terme $\gamma_n \cdot \frac{d\delta_n}{dt}$ représente <b>l'amortissement.</b> Cette valeur permet d'amortir les contacts trop intenses elle est répulsive au début de l'impact et attractive à sa fin. </p>

<a id="IIb"></a>
### II.b) Dynamique et schéma d'intégration temporel :

<p style="text-align: justify;"> En dehors des équations modélisant les efforts, si nous voulons <b>modéliser la dynamique de nos particules</b>, nous devons faire appel aux <b>Principe Fondamental de la Dynamique</b>. A partir du PFD de Newton, nous pouvons quantifier l'accélération d'un système via le bilan des actions extérieures qui s'appliquent sur celui-ci. </p>

<p style="text-align: justify;"> Néanmoins, dans le but de prévoir <b>l'évolution de la vitesse et de la position d'un grain</b>, il est nécessaire d'intégrer cette équation fondamentale. C'est pourquoi, afin d'intégrer temporellement ces équations nous devons faire appel à un "<b>schéma d'intégration temporel</b>". L'intégration temporelle est l'ensemble des techniques qui permettent de résoudre numériquement des <b>équations différentielles ordinaires</b>, qui sont des équations décrivant le comportement d'une variable au fil du temps. Ici le schéma est utilisé pour le PFD et les équations qui en découlent. </p>

<p style="text-align: justify;"> Une fois que nous avons choisi et appliqué un schéma d'intégration, nous pouvons prédire <b>l'évolution de la position et de la vitesse de chaque particule au fil du temps</b>, en fonction des <b>forces externes qu'elle subit.</b></p>

<p style="text-align: justify;"> <b>Le schéma d'intégration temporel utilisé dans notre projet est celui de Verlet</b>. Le schéma d'intégration temporel de Verlet est un excellent choix pour notre projet parce qu'il permet <b>une intégration précise des équations du mouvement</b> pour nos particules granulaires sur des pas de temps discrets. Ce schéma garantit une conservation rigoureuse de l'énergie, ce qui est crucial dans notre simulation, car la conservation de l'énergie est l'une des propriétés fondamentales des systèmes physiques. En outre, sa <b>simplicité</b> et sa <b>stabilité numérique</b> le rendent particulièrement adapté à la modélisation de grands systèmes de particules comme les silos à grains dans notre simulation. Nous pouvons décrire ce schéma étape par étape comme suit: </p>

<p style="text-align: justify;"> Le terme $\Delta t$ est le pas de temps de notre simulation. <b>C'est probablement le paramètre le plus important de la simulation</b>. En effet, plus un pas de temps est petit, plus le caractère discret du temps se rapprochera du caratère continu réel. <b>Un pas de temps faible implique une simulation plus longue</b> car impliquant un plus grand nombre de calcul pour une simulation à durée fixée. De plus, un grand nombre de calcul, résulte souvent en une occupation de mémoire (numérique) plus importante. Ce paramètre est en même temps une appréciation de la qualité de la simulation car central au cœur des calculs réalisés. <b>Par exemple</b>, généralement, l'une des premières erreurs que fait une personne est d'attribuer une trop grande <a href="#figure1"><b>raideur</b></a> pour un pas de temps pas assez faible. Une trop grande raideur conduit à une <b>réaction dynamique très forte</b> qu'il est nécessaire de suivre rapidement. Sans un pas de temps faible, il est impossible de suivre la forte réponse induite par une raideur importante. </p>

## III. Principales simplifications du modèle et choix d'implémentation

<a id="IIIa"></a>
### III.a) Simplifications :
- <p style="text-align: justify;"> Notre modèle ne considère pas la rotation des grains. Autrement dit, <b>nos grains ne 'roulent' pas, ils glissent</b>. Notre modèle se limite donc à deux degrés de liberté (l'abscisse et l'ordonné) soit une <b>modélisation en 2D</b>. Voici une liste des différents phénomènes physiques implémentés :</p>

    + Contact grain-grain, grain-paroi, grain-bac:
        + Résultante normale.
        + Amortissement.
        + Résultante tangentielle.
        + Glissement.
    + Autres:
        + Efforts à distance (gravite due au poids des grains)
        + Frottement de l'air (trainée aéronotique)  
  
- <p style="text-align: justify;"> Par ailleurs, si dans la réalité les granulés possèdent tous une forme singulière, pour ce projet, <b>nous avons uniquement considéré des grains sous forme de boules parfaites. Toutefois,</b> les rayons des grains varient de <a id="variation_rayon"></a>$\pm$20% par rapport à une valeur de référence. La masse associée aux grains et calculée via la masse volumique des grains. </p> 

- <p style="text-align: justify;"> Il existe d'autres simplifications physiques (utilisation du PFD, ...) et mécanique (solide indéformable, ...) fondamentales qui ne seront pas détaillées ici. </p>

<a id="voisinage"></a>  
- <p style="text-align: justify;"> En lisant attentivement <a href="#figure5"><b>l'algorithme</b></a> de notre programme, on peut y lire l'utilisation d'<b>une liste de voisinage</b>. En effet, lorsqu'on réalise des simulations numériques, le rendu final prend généralement <b>énormément de temps</b> à être finalisé. C'est pourquoi, pour gagner du temps sur nos calculs, on évite généralement de calculer toutes les variables entre tous les grains et toutes les parois à chaque instant de la simulation. A la place, on implémente <b>une liste de contact</b> qui attribue à chaque grain les contacts qu'il réalise avec son environnement (autres grains ou parois). Pour réaliser cette liste de contact, on utilise alors <b>une liste de voisinage propre à chaque grain</b> qui rend compte de l'environnement autour de l'objet. Il existe plusieurs manières de concevoir des listes de voisinages, celle retenue dans notre projet se nomme <b>la méthode de partition du conteneur</b>: </p>

<a id="figure8"></a>

<p style="text-align: justify;">Cette méthode implique la réalisation d'<b>une grille discrétisant l'espace selon un pas d'espace</b>. Chaque grain appartient alors à une case de la grille. Néanmoins, il est aussi nécessaire de régulièrement mettre à jour, à la fois la grille ainsi que la liste de contact, selon un intervalle de temps multiple $n_v$ (défini comme paramètre) du pas de temps. Une explication de son implémentation complexe sera détaillée plus loin.</p>

- <p style="text-align: justify;">
    Comme nous venons de l'indiquer, les grains possèdent tous <b>un rayon aléatoire</b> variant de plus ou moins 20% autour d'un rayon abitrairement choisi. Pour notre simulateur <b>nous avons choisi un rayon de référence de 5 $mm$, ainsi qu'une masse volumique de 770 $kg / m^3$</b>. Autrement dit, pour tous les grains, leur rayon sera compris tel que: $ 4 < r < 6,$ en $mm$ et la masse correspondante: $m=\frac{4}{3} \pi r^3 \rho \space(kg)$, avec $p$ la masse volumique et $r$ le rayon. Ces deux valeurs ne sont pas anodines, elles impactent grandement la simulation et il est important de les choisir avec précaution. C'est pourquoi, comme vous le comprendrez dans la suite de ce compte-rendu, nous ne les avons pas inscrites dans l'interface utilisateur à disposition. Pour comparer nos résultats et comprendre l'impact des changements que nous opérions sur notre simulateur, il était important de <b>considérer le même type de granulé.</b>
</p>

### IV.a) Interface utilisateur:
<p style="text-align: justify;"> 
    Le fichier "Fenetre_paremetre" est <b> une interface graphique</b> réalisée via la bibliothèque pyhon <b>tkinter.</b> Elle autorise l'utilisateur à définir à la fois <b>la situation initiale de l'écoulement granulaire (nombre de grains, position des parois, etc.) de même que les paramètres de la simulation <a href="#ref_relation">précédemment discutés</a>.</b> Voici les différents paramètres qu'il est possible de définir à travers cette interface:
<ul>
    <li> Coeff. dir.: coefficient directeur des deux droites modélisant les parois du silo. </li>
    <li> Ord. origine: l'ordonnée à l'origine des deux droites. </li>
    <li> Debut Trou: ordonnée $y=cste$ à partir de laquelle les grains tombent.</li>
    <li> Durée simulation: le durée de la simulation souhaitée. </li>
    <li> <a id="bac_equation"> </a>Hauteur bac: ordonnée $y=cste$ de la droite modélisant le bac de réception des grains. </li>
    <li> Hauteur grain: ordonnée du premier grain à placer dans le silo (le placement des autres grains se fait automatiquement). </li>
    <li> Nombre de grains: le nombre de grains de la simulation. </li>
</ul>
L'interface propose aussi deux "checkbox" permettant de faire un choix binaire (oui/non, vrai/faux, 1/0):
<ul>
    <li> Afficher les grains: permet d'afficher dans l'interface le placement intial des grains. </li>
    <li> Activer la physique du bac: permet d'activer le contact/l'intéraction avec le bac, si cette case est cochée le bac se comportera comme une paroi, sinon les grains arrivant dessus ne seront plus mis à jour et ils deviendront totalement inertes, ces grains ne bougent plus et n'intéragissent plus avec leur environnement. </li>
</ul> 
</p>

<p style="text-align: justify;">
    <b> Lors du démarrage de notre simulateur </b> cette interface apparaît. L'utilisateur définit les paramètres souhaités puis appuie sur le bouton en bas de la fenêtre "Enregistrer". De cette manière, <b>les paramètres sont enregistrés sous la forme d'un fichier csv</b> (parametres.csv) et sont utilisés comme <b>paramètres initiaux</b> lors du lancement suivant, sinon ils ne le sont pas. La simulation se lance après la fermeture de l'interface avec les derniers paramètres inscrit à travers celle-ci.
</p>

### IV.b) Programme principal :

<a id="IVb1"></a>
### IV.b.1) Définition:
<p style="text-align: justify;">
    Dans un premier temps, nous allons expliquer en profondeur comment certains <b>tableaux</b> et certaines <b>variables</b>, qui peuvent être complexe à appréhender au premier abord, sont définis.
</p>

<p style="text-align: justify;">
    Au sein de la partie "DEFINITION" le programme réalise toutes <b>les définitions de variables, de tableaux de données ainsi que de script permettant d'initialiser la simulation.</b> A l'intérieur de cette sous-partie, nous définissons tout ce qui touche à/aux:
</p>

- grains: nombre de grains, rayon, masse volumique, masse, raideurs, coefficient de trainée, coefficient d'amortissement.
- temps: temps actuel, indice du temps, pas de temps, durée de la simulation, nombre de (pas de) temps.
- espace: limites haut, gauche, droite, coefficient de frottement, pas d'espace, nombre de cases et la grille.
- tableaux de données: position, vitesse, vitesse demi pas, acceleration, contact, allongement, mise à jour et un tableau masse_passee pour le débit massique.
- silo: hauteur du bac, largeur du bac, coefficient directeur, etc...
- positionnement des grains: script positionnant les grains au temps initial 0.

<p style="text-align: justify;">
    Commençons par les définitions sur les variables liées aux grains. <b>Les variable de rayon et de masse sont en réalité des tableaux.</b> En effet, comme nous l'avons <a href="#variation_rayon">précisé précédemment</a> le rayon des grains varie de plus ou moins 20% par rapport à une valeur moyenne choisie. Le choix du rayon au sein de cette intervalle est totalement <b>aléatoire.</b> C'est pourquoi, chaque grain possède un rayon qui lui est propre. Pour la masse des grains elle est calculée grâce à la <b>masse volumique</b> inscrite dans notre programme. Ces deux variables sont donc des tableaux qui contiennent pour un grain d'indice $i$ les valeurs de rayon et de masse qui lui sont propres.
</p>

<p style="text-align: justify;">
    Qu'en est-il des variables définissant l'espace ? La signification des variables "limites" n'est pas évidentes. Le principal interêt de cette sous-section est l'implémentation de la <a href="#figure8"><b>grille discrétisant l'espace</b></a>.
</p>
<a id="explicationgrille"></a>

- <p style="text-align: justify;"> A l'instar de la <b>discrétisation temporelle</b>, si nous voulons discrétiser un espace il nous faut en définir <b>les limites.</b> La où le temps commence à 0$s$ et s'arrête après la durée de simulation choisie, on définit, dans le plan, <b>un carré délimité par 4 côté d'équations: $x_{gauche} = cste$, $x_{droite} = cste$, $y_{bas} = cste$ et $y_{haut} = cste$.</b> Ces limites ont été choisies <b>arbitrairement</b>. Autrement dit, il est tout à fait possible de faire dysfonctionner le programme si un grain dépasse ces bornes. Le carré est défini selon les points du plan: $(-1.5,-1), \space (1.5, 2)$ (les valeurs sont en mètre). </p>
- <p style="text-align: justify;">Au même titre que nous avons besoin d'un pas de temps, nous avons ici besoin d'<b>un pas d'espace</b>, duquel nous pouvons définir la dimension d'une case. Le pas d'espace défini est de <b>5 fois le rayon de référence</b> désigné.</p> 
- <p style="text-align: justify;">A l'aide du pas et des limites d'epaces, nous pouvons définir le nombre de cases contenues dans la section du plan définie: $ nb_{cases selon x} =  \lfloor\frac{(limite_{droite} - limite_{gauche})}{c}\rfloor +2$ <b>avec c, le pas d'espace</b>. On ajoute $+2$ pour éviter certaines aberrations aux limites. On fait pareil pour le nombre de cases selon l'ordonnée. </p> 
- <p style="text-align: justify;">Désormais nous pouvons définir la grille. Pour définir la grille, il a d'abord été imaginé un dictionnaire python qui associerais à la case $(i, j)$, la liste des grains y appartenant. <b>Pour des raisons techniques, cette solution n'a pas été retenue</b>. Une autre option, plus brutale, a été imaginée. La grille est <b>un tableau 3D</b> contenant les coordonnées des cases $(i, j)$ comme entrée et associant un 0 (False) ou un 1 (True) selon l'indice $k$ du grain. Pour toutes les cases, nous associons donc au tableau 2D une <b>liste de la taille du nombre de grain</b> contenant des 0 et des 1. En d'autres termes, chaque grain ne peut avoir qu'un seul "1" dans une seule case du tableau correspondant à son appartenance à cette case de la grille. Cette méthode est <b>très couteuse</b>, mais son choix est justifié. La raison technique nous ayant poussé à ce choix est l'utilisation d'une bibliothèque python: <a id="numba0"></a><b>Numba. Numba est une bibliothèque Python qui permet d'accélérer l'exécution de code en le compilant just-in-time (JIT)</b> pour tirer parti de l'optimisation des performances du processeur. Elle permet ainsi d'<b>améliorer les performances des calculs numériques et scientifiques en Python</b> de manière simple et efficace. Cependant, la manipulation de dictionnaires avec la biliothèque Numba n'est pas très aisée. C'est pourquoi, afin de garder une utilisation globale de cette bibliothèque dans le reste du programme python nous avons préféré choisir cette méthode.
</p>

    <figure style="text-align: center; padding: 20px">

    <figcaption style="border: 1px solid black; padding: 10px; display: inline-block;"><b>Fig.10 - Code implémentant les variables d'espace</b></figcaption>
</figure>

> Remarque: lorsqu'il est inscrit "app." cela signifie que la variable est récupérée depuis le fichier définissant l'interface utilisateur.
    

<p style="text-align: justify;"> <a id="tableaux"></a>
    A présent, présentons les autres tableaux de données. Les tableaux de position, vitesse, etc. sont triviaux, ils contiennent pour <b>chaque temps de la simulaton</b> et pour <b>chaque grain</b>, les composantes des vecteurs position, vitesse, etc... <b>Les tableaux "mise_a_jour", "CONTACT", "ALLONGEMENT" et "GRAINS_PASSES" sont différents</b>:
</p>

- <p style="text-align: justify;">Le tableau "mise_a_jour" est un tableau de la taille du nombre de grain. Pour chaque grain il attribue un 1 (True) ou un 0 (False) <b>décrivant si le grain doit être mis à jour ou non</b>. Ce tableau est lié à la <a href="#gui">"checkbox"</a> d'activation de la physique du bac de réception. Lorsqu'un grain arrive sur le bac de reception est que la "checkbox" n'est pas coché, alors le grain vera sa valeur passer de True à False et sera totalement inerte sans aucun mouvement ni aucune intéraction.</p> 

<a id="precedcommentaire"></a>
- <p style="text-align: justify;">Le tableau "CONTACT" à une construction similaire. L'objectif de ce tableau est de <b>stocker les contacts qu'il y a lieu</b>. C'est un <b>tableau 3D de taille (nombre de grain, nombre de grains + 3, 2)</b>. Ce tableau associe à chaque paire de grains deux valeurs booléennes (True/False). La première indique s'il y a <b>contact</b> et la seconde indique s'il y a <b>glissement.</b> Nous ajoutons un +3 à "nombre de grains + 3" afin de prendre en compte les parois et le bac. Les indices $i$ appartenant à l'intervalle $[0:nb_{grains}[$ sont les indices des grains, tandis que les trois derniers indices $nb_{grains}$, $nb_{grains+1}$ et $nb_{grains+2}$ correspondent, respectivement, à <b>la paroi gauche, droite et le bac.</b> On peut donc résumer qu'à un couple $(k,i)$, avec $k$ l'indice d'un grain et $i$ l'indice d'un grain ou un obstacle, est associé un couple de valeur booléen. Le couple booléen transcrivant le caractère du contact, existant ou non et glissant ou non.</p>

- <p style="text-align: justify;"> Le tableau "ALLONGEMENT" a la même configuration que le tableau "CONTACT". La différence entre les deux et qu'au lieu de stocker des booléens, ce tableau contient <b>le couple de valeurs indiquant l'allongement du contact.</b> La première valeur est <b>l'allongement normal $\delta_n$</b>, la seconde est <b>l'allongement tangentiel $\delta_t$.</b> Ce sont des distances elles donc stockées comme des flottants sur 8 octets. </p>

- <p style="text-align: justify;"> Le tableau "MASSE_PASSEE" est le tableau nécessaire au calcul du débit de l'écoulement granulaire. Sa taille est de "nb_temps" et, pour chaque indice du temps, il associe la masse totale qui est déjà passée par l'ouverture du silo.</p>
    
    
<p style="text-align: justify;">
    Nous pouvons également nous intéresser à la définition du silo:
    </p>

- <p style="text-align: justify;"> Le bac de réception est défini par la variable <a href="#bac_equation"> "hauteur_bac" </a>. Les variables de largeur n'ont aucun impact sur l'intéraction grain-bac. Si un grain se trouve en-dessous de la droite d'hauteur "hauteur_bac" alors elle est en concact avec ce dernier. </p>
- <p style="text-align: justify;"> En revanche, la définition des deux droites symétriques formant les parois du silo est faites à partir de <b>l'équation mathématiques d'une droite dans le plan: $y = Ax + C$ avec $A$ et $C$ deux valeurs réelles.</b> Un grain se trouve en contact avec l'une des deux parois si <b>la distance orthogonale qu'il la sépare de la paroi est inférieure ou égale à 0</b>: $d(\mathrm{M}, (d)) = \mathrm{MM}_h = \frac{|ax_M+by_M+c|}{\sqrt{a^2+b^2}} \leq 0$. Il faut faire <b>attention</b>, dans la formule précédente $A$, $B$ et $C$ sont les coefficients de la droite $(d)$ définit par l'équation suivante: $ax + by + c = 0$, étant donné la précédente forme de l'équation de droite que nous avons choisie, il faut souligner que pour nous $b = 1$, $a=-A$ et $c=-C$  </p> 

> Remarque: la seule différence entre les deux parois(gauche et droite) et la valeur du coefficient A qui est opposé: $A_{gauche} = -A_{droite}$

    </p>

<p style="text-align: justify;"> Ce qui est réjouissant avec ce genre de projet est <b>l'obtention d'une simulation concrète.</b> Une balle qui rebondit ou des grains de sables qui forment des monticules, les résultats obtenus nous sont familiers. ### V.a) Historique des résultats et difficultés rencontrées : 
<p style="text-align: justify;"> Il est intéressant de remarquer que les résultats obtenus ont été très linéaires. Dans un premier temps nous avons simplement essayé de créer <b>les intéractions de rebond entre les grains.</b> Ensuite l'implémentation silo est rapidement devenue une necessité. Nous avons précisé les intéractions, rencontré de nombreux problèmes de programmation, nous avons effacé puis réécrit, puis reculé, etc. </p>

<p style="text-align: justify;"> Les premières et les dernières difficultés mais aussi celles qui nous ont suivies tout du long de la réalisation de notre projet sont les suivantes: <b>la connaissance, la consistance et l'organisation.</b> Encore à la fin du projet, des erreurs d'inattention et de communication nous ont values de très mauvais résultats ainsi que d'importantes pertes de temps. De plus, savoir quoi faire, comment modéliser les intéractions, où trouver les équations physiques et les informations nécessaires n'a pas été une mince affaire. Doublés de notre manque de consistance, ces erreurs nous ont obligées à très régulièrement revenir sur nos pas. Ces trois majeures complications se sont toujours retrouvées dans le reste des autres obstacles que nous avons dû surmonter. </p>

<p style="text-align: justify;"> En continuant le développement de l'application, les calculs <b>s'allourdissent et le temps nécessaire pour effecuter la simulation grandit.</b> Pour parer à ce problème, nous allons faire un changement qui nous causera beaucoup de soucis. Nous l'avons déjà mentionné dans ce rapport, mais nous avons utilisé la bibliothèque <a href="#numba0"><b>Numba</b></a>. Si cette bibliothèque permet de <b>grandement accélerer la vitesse d'éxecution de notre programme</b>, elle est aussi très limitée par ses applications. Numba est un compilateur JIT (Just-In-Time) pour Python, développé par Anaconda, qui permet d'accélérer le code Python en le compilant en bytecode de bas niveau, optimisé pour notre ordinateur spécifique au moment de l'exécution. Il est particulièrement utile pour <b>les tâches de calcul lourd telles que les boucles, les tableaux de grande taille et les fonctions numériques.</b> Notre utilisation de numba s'est limitée au décorateur "@njit". Le terme "nJIT" est un acronyme pour "no Python JIT". Cela signifie que lors de l'utilisation du décorateur "@njit", Numba génère du code qui ne fait pas appel à l'interpréteur Python. Cela signifie que certaines fonctionnalités de Python qui dépendent de l'interpréteur deviennent inutilisable. En termes de limitations, certaines des principales sont : </p>

-  <p style="text-align: justify;">Numba ne supporte pas tout le langage Python. Par exemple, nous ne pouvons pas utiliser des fonctionnalités telles que les générateurs ou certains types de données haut-niveau comme les dictionnaires.</p>
-  <p style="text-align: justify;">Numba peut ne pas offrir d'améliorations de performance significatives pour tous les types de code. En particulier, il excelle dans les tâches de calcul lourd, mais pour le code qui est principalement lié à l'E/S (entrée/sortie) ou qui fait beaucoup d'appels à des bibliothèques externes, l'utilisation de Numba peut ne pas offrir d'amélioration notable.</p>
-  <p style="text-align: justify;">Bien que Numba supporte un sous-ensemble de NumPy, il n'y a pas une couverture complète. Certains fonctions et certaines fonctionnalités peuvent ne pas être disponibles.</p>

<p style="text-align: justify;"> En résumé, l'utilisation de Numba demande de <b>reconstruire entièrement notre implémentation sous python du simulateur</b>, multipliant les erreurs à corriger. C'est pourquoi, dans notre code, il y a exclusivement des tableaux numpy. Son utilisation justifie aussi notre choix à utiliser des tableaux de grandes tailles et des calculs très lourds. </p>

<a id="Vb"></a>
### V.b) Critique globale du simulateur : 
Points positifs:
- Un simulateur qui dépeint de manière satifaisante ce que l'on pourrait attendre d'un écoulement granulaire en 2D.
- L'implémentation de phénomènes physiques complexes (notamment le glissement et l'allongement tangentiel).
- Un outil facile et agréable à utiliser et qui donne des résultats directement (pas d'enregistrement, affichage instantané).
- Un programme facile à lire et correctement documenté (nombreux commentaires dans le code).
- De belles animations.

Points négatifs:
- Un programme dont les temps calculs peuvent très rapidement augmenter.
- Une sensibilité accrue au niveau des paramètres et une définition parfois injustifiée de ces derniers(par exemple le coefficient de frottement ou de trainée qui ont été lourdement modifié au cours du projet).
- Des grains qui ne roulent pas.
- La loi empirique du débit de Beverloo, même très modifiée, ne correspond pas aux valeurs retrouvées numériquement. De plus, ces calculs de débits deviennent totalement aberrants lorsqu'on écarte les parois du silo. Ces aberrations sont dues au frottement visqueux que les grains entretiennent avec les parois du silo (coefficient de frottement dans le projet).

Améliorations futures:
- Il serait très intéressant de reprogrammer entièrement le simulateur pour une lecture encore plus claire et des optimisations avancées (programmation orienté objet, utilisation des cores graphiques cuda, etc.).
- L'implémentation du roulement.
- Une définition plus poussée des paramètres permettant une adaption plus ample de la simulation.
- Une interface graphique complète affichant, non pas dans la console le reste des paramètres, mais autorisant une paramètrisation complète et un suivi entier du déroulement de l'exécution.
- Résolution de tout les phénomènes induisant un plantage de la simulation. 

<a id="VI"></a>
## VI. Conclusion
---
Ce projet a été un voyage fascinant à travers le domaine de la simulation granulaire. Malgré les nombreuses difficultés rencontrées, que ce soit en matière <b>de compréhension théorique ou d'implémentation pratique</b>, le résultat est particulièrement satisfaisant. Nous avons réussi à créer un simulateur qui représente de manière convaincante l'écoulement granulaire en 2D.

Il convient de noter que les résultats obtenus ont été très linéaires, évoluant au fil du temps à mesure que nous apportions <b>des corrections et des améliorations.</b> Nous avons également tiré des leçons précieuses de nos erreurs, en particulier en ce qui concerne l'importance de la consistance, de l'organisation et de la communication dans le cadre d'un projet de cette ampleur.

Notre simulateur n'est pas sans défauts. Il a <b>une sensibilité accrue à certains paramètres</b> et <b>son temps de calcul peut rapidement augmenter.</b> De plus, il y a encore des améliorations à apporter, notamment l'implémentation du <b>roulement des grains</b>, une meilleure définition des paramètres et une interface utilisateur plus complète.

Cependant, le simulateur a également de nombreux points forts. Il modélise des phénomènes physiques complexes comme <b>le glissement et l'allongement tangentiel</b>, fournit des animations de haute qualité et est facile à utiliser. De plus, le code est bien documenté, ce qui facilite sa compréhension.

En conclusion, ce projet a été une occasion d'approfondir nos connaissances en physique et en informatique, tout en créant un outil utile. En effet, l'un des avantages de ce type de projet est qu'il peut toujours être <b>amélioré et étendu</b>, offrant ainsi d'innombrables possibilités pour l'avenir.
